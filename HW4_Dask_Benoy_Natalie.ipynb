### Task 9f a

```python
# compute ratio of non-cancelled flights to total flights, by origin
ddf_cancelled = ddf.loc[ddf['Cancelled'] == 0, ['Cancelled', 'Origin']].groupby('Origin').agg('count').compute()
ddf_not_cancelled = ddf.loc[ddf['Cancelled'] == 1, ['Cancelled', 'Origin']].groupby('Origin').agg('count').compute()
ddf_total = ddf_cancelled + ddf_not_cancelled
Ratio = ddf_cancelled / ddf_total
print(Ratio)

# it ain't pretty but it DOES work
```

## Task 1

```python
# generate random samples of data
from sklearn.datasets import make_classification
X, y = make_classification(n_samples = 10000, n_features = 4, random_state = 0)

print(X[:8])
y[:8]
```

## Task 2

```python
# fit SupportVector classifier
from sklearn.svm import SVC

estimator = SVC(random_state = 0)
estimator.fit(X, y)

print(estimator.support_vectors_[:4])

estimator.score(X, y)
```

## Task 3

```python
# hyperparameter tuning
estimator = SVC(C = 0.00001, shrinking = False, random_state = 0)
estimator.fit(X, y)
print(estimator.support_vectors_[:4])
estimator.score(X, y)
```

## Task 4

```python
# apply grid search on one core
from sklearn.model_selection import GridSearchCV
estimator = SVC(gamma = 'auto', random_state = 0, probability = True)

param_grid = {
    'C': [0.001, 10.0],
    'kernel': ['rbf', 'poly'],
}

grid_search = GridSearchCV(estimator, param_grid, verbose = 2, cv = 2)

# see how long it takes
%time grid_search.fit(X, y)
```

## Task 5

```python
# compare timing with ScikitLearn parallelism
grid_search_MC = GridSearchCV(estimator, param_grid, verbose = 2, cv = 2, n_jobs = -1)

# see how long it takes
%time grid_search_MC.fit(X, y)
```

```python
# find optimal hyperparameters
grid_search_MC.best_params_, grid_search_MC.best_score_
```

## Task 6

```python
# parallelism with dask
import joblib

param_grid = {
    'C': [0.001, 0.1, 1.0, 2.5, 5.],
  #  'kernel': ['rbf', 'poly', 'linear'],
  #  'shrinking': [True, False],
}

grid_search_Dist = GridSearchCV(estimator, param_grid, verbose = 2, cv = 5, n_jobs = -1)

# see how long it takes
# spoiler alert: a long time

%time
with joblib.parallel_backend('dask'):
    grid_search_Dist.fit(X, y)
```

```python
# view grid search results as data frame
pd.DataFrame(grid_search_Dist.cv_results_).head()
```

```python
# find optimal hyperparameters, score
grid_search_Dist.best_params_, grid_search_Dist.best_score_
```

```python
# obtain summary of other performance measures
y_pred = grid_search_Dist.predict(X)

from sklearn.metrics import classification_report
print(classification_report(y, y_pred))
```
