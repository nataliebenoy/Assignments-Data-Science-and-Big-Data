```python
# add a new column
popDF['2019_pctg'] = (popDF['y2019'] - popDF['y2018']) / popDF['y2018']
```

### Answer 11 i

```python
# compute and add a new column in popDF that contains the compound annual growth rate of population
# between 2014 and 2019. Call this new column ‘2019_5cagr’. Show 5 random rows of popDF.

popDF['2019_5cagr'] = np.round((((popDF['y2019'] / popDF['y2014']) ** (1/5)) - 1), 4)

popDF.sample(5)
```

```python
# more functions
popDF[['y2019']].apply(lambda x : np.exp((x - x.mean()) / x.std())).head()

def expstandardize(x):
    return np.exp((x - x.mean()) / x.std())

popDF[['y2019']].apply(expstandardize).head()
```

### Answer 11 ii

```python
# compute a new column in popDF that contains the coefficient of variation for each state for the population
# in years 2010-2019. Use DF.apply() for this task. Name this column ‘coeffofvar’.
# Display the first 5 values of only this new column (with index).

popDF['coeffofvar'] = popDF[['y2010', 'y2011', 'y2012', 'y2013', 'y2014', 'y2015', 'y2016', 'y2017', 'y2018', 'y2019']].apply(
    lambda x : x.std() / x.mean(), axis = 1)

popDF[['coeffofvar']].head(5)
```

```python
# more aggregate operations
popDF[['y2018', 'y2019']].agg(
    ['count', 'sum', 'min', 'max', 'mean', 'var', 'skew', 'kurt', 'median', 'idxmax', 'idxmin'])
```

```python
# even more aggregate operations
popDF[['y2018', 'y2019']].agg({'y2019' : ['min', 'max'], 'y2018' : ['var', 'std']})
```

```python
# even more aggregate operations
popDF[['y2018', 'y2019']].agg({'y2019' : ['min', 'max'], 'y2018' : ['var', 'std', np.min, np.max]})
```

```python
# using group by
popDF[['Region', 'y2010', 'y2019']].groupby(['Region']).mean().sort_values(['Region'])
```

```python
# more group by
popDF.groupby(['Region'])[['y2010', 'y2019']].agg(['min', 'idxmin', 'max', 'idxmax', 'mean'])
```

### Answer 11 iii

```python
# compute a new column in popDF that contains the mean of the population in year 2019 grouped by region
# (Midwest, Northeast, South, West). Call this new column ‘mean2019popbyReg’.
# Print the first 5 rows of only this column (with index) as a DataFrame.

popDF['mean2019popbyReg'] = popDF.groupby(['Region'])[['y2019']].transform('mean')

popDF[['mean2019popbyReg']].head(5)
```
